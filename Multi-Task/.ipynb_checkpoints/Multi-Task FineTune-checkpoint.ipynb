```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D

import qgrid
```

# Sort by Testing F1 score

```python
df = pd.read_csv("model_results/model_scores.csv")
```

```python
df = df.sort_values('Testing F1', ascending=False)
df.index = df.Model
models = list(df.Model)
```

```python
df['Batch size'] = df.Model.apply(lambda a : a.split('_')[2][1:])
df['Dropout'] = df.Model.apply(lambda a : a.split('_')[3][1:])
df['LSTM size'] = df.Model.apply(lambda a : a.split('_')[4][4:])

df.index = pd.MultiIndex.from_arrays([df['Batch size'], df['Dropout'], df['LSTM size']])
del df['Model']
del df['Batch size']
del df['Dropout']
del df['LSTM size']

qgrid.show_grid(df)
```

```python
df
```

# Models and F1 score per tasks

```python
f1s = list(df['Testing F1'])
t1s = []
t2s = []
t3s = []


for m in models:

    task1 = []
    task2 = []
    task3 = []
    val_f1 = []

    # RETRIEVE MULTI-TASK F1 SCORES
    with open ("model_results/{0}/{0}.txt".format(m), "r") as myfile:
        for line in myfile:
            if "val_f1" in line:
                x = line.split('val_f1')[1].split(': ')
                val_f1.append(float(x[1].split(' ')[0]))
                x = x[2:]
                task1.append(float(x[0].split(' ')[0]))
                task2.append(float(x[1].split(' ')[0]))
                task3.append(float(x[2].split('\n')[0]))

    x = np.argmax(val_f1)
    t1s.append(task1[x])
    t2s.append(task2[x])
    t3s.append(task3[x])
```

```python
df = pd.DataFrame(list(zip(models, f1s, t1s, t2s, t3s)), columns=["Model", "F1", "Task 1", "Task 2", "Task 3"])
qgrid.show_grid(df)
```

```python
df
```

# Best Model investigation

```python
for best_model in models:
    print(best_model)
    
    task1 = []
    task2 = []
    task3 = []
    val_f1 = []

    # RETRIEVE MULTI-TASK F1 SCORES
    with open ("model_results/{0}/{0}.txt".format(best_model), "r") as myfile:
        for line in myfile:
            if "val_f1" in line:
                x = line.split('val_f1')[1].split(': ')
                val_f1.append(float(x[1].split(' ')[0]))
                x = x[2:]

                task1.append(float(x[0].split(' ')[0]))
                task2.append(float(x[1].split(' ')[0]))
                task3.append(float(x[2].split('\n')[0]))



    # PLOT            
    plt.figure(figsize=(20,7))

    line1, = plt.plot(task1, 'k', label="Task 1")
    line2, = plt.plot(task2, 'b', label="Task 2")
    line3, = plt.plot(task3, 'g', label="Task 3")

    line4, = plt.plot(val_f1, 'r', label="Mean F1", linestyle='--')
    plt.axvline(x=np.argmax(val_f1), color='black', label='Best epoch')

    plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})

    axes = plt.gca()
    axes.set_ylim([0.7,1])
    axes.set_xlim([0,25])

    plt.title("Tasks F1 scores per epoch")
    plt.xlabel("Nbr epoch")
    plt.ylabel("F1 score")

    plt.show()
    
    print('')
    print('-----------------------')
    print('')
```
